### Joint plot from two gamma distributions
#### We sample from these two distributions using Monte Carlo sampling
#### We calculate the posterior distribution of $\theta_{1} > \theta_{2}$ and $\theta_{1} / \theta_{2}$

```python
# SciPy defines the parameters for parameters in a different manner compared to the earlier definition
# the scale parameter is 1/b
def joint_sample(params1=(219,1.0/112), params2=(68,1.0/45)):
    nsize = 5000
    a1 = random.gamma(params1[0], params1[1], nsize)
    a2 = random.gamma(params2[0], params2[1], nsize)
    
    a1_greater_than_a2 = 0
    a1_divided_by_a2 = [0] * nsize
    
    for i in range(nsize):
        
        if(a1[i] > a2[i]):
            a1_greater_than_a2 += 1
        
        a1_divided_by_a2[i] = a1[i] / a2[i]
            
    print("Approximate mean from sampling for distribution a1 greater than a2", a1_greater_than_a2 / nsize)
    fig = ff.create_distplot([a1_divided_by_a2], group_labels=['a1/a2'], show_rug=False, bin_size=0.1)
    fig.show()
    
        
joint_sample()
```
